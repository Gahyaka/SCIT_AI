Pengertian Statistik Deskriptif

Statistik deskriptif adalah cabang statistik yang digunakan untuk mengumpulkan, menyajikan, dan menganalisis data sehingga dapat memberikan gambaran atau ringkasan informasi tanpa melakukan generalisasi atau penarikan kesimpulan untuk populasi yang lebih luas.

Tujuan Statistik Deskriptif

1. Menyajikan data dalam bentuk yang lebih mudah dipahami.
2. Merangkum data menjadi informasi inti.
3. Membantu mengidentifikasi pola atau tren awal pada data

Ukuran Pemusatan (Measures of Central Tendency)

Mean (rata-rata)\
Median (nilai tengah)\
Mode (nilai yang paling sering muncul)

```python
import pandas as pd

# Contoh dataset sederhana
data = {
    "Pendapatan": [5, 7, 8, 6, 9, 12, 15, 7, 9, 10],
    "Harga": [100, 120, 130, 115, 140, 160, 200, 110, 135, 150],
    "Kepuasan": [3, 4, 4, 3, 5, 5, 4, 3, 4, 5]
}

df = pd.DataFrame(data)

print("Mean:", df.mean())
print("Median:", df.median())
print("Modus:", df.mode()) 
```

Coefficient of variation

DISTRIBUSI

```python

import numpy as np
from scipy import stats
import pandas as pd

# Baca data survei_tinggi_badan.txt dengan numpy loadtxt
tinggi_badan = np.loadtxt("data_survei_tinggi_badan.txt", skiprows=1)

#berbeda dengan describe pandas
df= pd.DataFrame(tinggi_badan, columns=["Tinggi Badan"])
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Buat figure dan axes (ax) dengan ukuran 8x6 inci
fig, ax = plt.subplots(figsize=(8, 6))

# Plot histogram frekuensi (jumlah absolut) pada axes pertama (ax)
sns.histplot(
    x=tinggi_badan,                          # data tinggi badan
    binwidth=1,                              # lebar tiap bin 1 cm
    binrange=(tinggi_badan.min()-0.5, tinggi_badan.max()+0.5),  # rentang bin
    ec="w",                                  # edgecolor (garis pinggir bar) warna putih
    ax=ax                                    # tampilkan pada axes pertama
)

# Tambahkan sumbu y kedua (twin y-axis) pada axes yang sama
ax1 = ax.twinx()

# Plot histogram frekuensi relatif (probabilitas) + KDE (kurva distribusi)
sns.histplot(
    x=tinggi_badan,                          # data tinggi badan
    stat="probability",                      # sumbu y menampilkan probabilitas (frekuensi relatif)
    binwidth=1,                              # lebar bin 1 cm
    binrange=(tinggi_badan.min()-0.5, tinggi_badan.max()+0.5),  # rentang bin
    kde=True,                                # tampilkan kurva KDE (estimasi distribusi)
    ec="w",                                  # garis tepi bar putih
    ax=ax1                                   # tampilkan pada axes kedua
)

ax.set_xlabel("Tinggi badan (cm)", fontsize=12)
ax.set_ylabel("Frekuensi", fontsize=12)
ax1.set_ylabel("Frekuensi relatif\n(Probabilitas)", fontsize=12)
ax1.grid(axis="y")
plt.tight_layout()
plt.show()
```

Histogram Kumulatif Data Survei

Untuk membuat histogram kumulatif berikut dengan fungsi distribusi tidaklah rumit,kita akan mengisikan parameter cumulative menjadi True pada kedua penggunaan method .histplot() yang telah diketikkan sebelumnya.

```python
# Buat figure sebagai canvas dengan ukuran 8 in x 6 in
fig, ax = plt.subplots(figsize=(8, 6))
# plotkan histogram untuk frekuensi 
sns.histplot(x=tinggi_badan, binwidth=1, binrange=(tinggi_badan.min()-0.5, tinggi_badan.max()+0.5), cumulative=True, ec="w", ax=ax)
# buat sumbu berikutnya agar frekuensi relatif dapat dijadikann pada satu plot
ax1 = ax.twinx()
# plotkan histogram untuk frekuensi relatif
sns.histplot(x=tinggi_badan, stat="probability", binwidth=1, binrange=(tinggi_badan.min()-0.5, tinggi_badan.max()+0.5), cumulative=True, kde=True, ec="w", ax=ax1)
# set label
ax.set_xlabel("Tinggi badan (cm)", fontsize=12) 
ax.set_ylabel("Frekuensi kumulatif", fontsize=12)
ax1.set_ylabel("Probabilitas kumulatif", fontsize=12)
ax1.grid(axis="y")
plt.tight_layout()
plt.show()
```

Persamaan luas pdf dan cdf



Menampilkan Dua Kurva: PDF dan CDF


```python
# Buat figure dan dua subplot (1 baris, 2 kolom) dengan ukuran canvas 10x5 inci
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot PDF (probability density function) pada subplot pertama (kiri)
sns.kdeplot(x=tinggi_badan, lw=3, ax=axs[0])

# Plot CDF (fungsi distribusi kumulatif dari KDE) pada subplot kedua (kanan)
sns.kdeplot(x=tinggi_badan, cumulative=True, lw=3, ax=axs[1])

# Plot ECDF (empirical CDF dari data asli) pada subplot kedua juga
sns.ecdfplot(x=tinggi_badan, lw=2, ax=axs[1])

# Set label sumbu-y subplot pertama
axs[0].set_ylabel("Probabilitas", fontsize=12)

# Loop: beri label sumbu-x dan garis grid pada kedua subplot
for ax in axs:
    ax.set_xlabel("Tinggi badan (cm)", fontsize=12)
    ax.grid(axis="y")

# Set label sumbu-y subplot kedua (yang terakhir di-loop)
ax.set_ylabel("Probabilitas kumulatif", fontsize=12)

# Tambahkan legenda pada subplot kedua
ax.legend(["cdf", "ecdf"], fontsize=12)

# Rapatkan layout agar tidak tumpang tindih
plt.tight_layout()

# Tampilkan plot
plt.show()

print("Plot diatas menjelaskan Data sampel tinggi badan ini memiliki distribusi mendekati distribusi normal seperti yang dinyatakan oleh teorema limit tengah")
```

Menentukan Probabilitas 

 Sekarang saatnya kita akan menentukan nilai probabilitas dan probabilitas kumulatif data terdistribusi normal jika diketahui nilai rata-rata dan standar deviasinya,

Penggunaan pdf dan cdf berdasarkan fungsi distribusi normal akan memudahkan masing-masingnya dalam menentukan nilai probabilitas dan probabilitas kumulatif. Pustaka scipy menyediakan method untuk menentukan nilai probabilitas dan probabilitas kumulatif yang dapat diakses melalui .norm() dari scipy.stats.

Untuk menentukan besar probabilitas dan probabilitas kumulatif dapat digunakan method berikut

Probabilitas:

norm.pdf(x, loc=0, scale=1)

Probabilitas kumulatif:

norm.cdf(x, loc=0, scale=1)

keduanya memiliki parameter yang sama dengan nilai parameter default untuk distribusi normal standar

x: bilangan (int, float) atau vektor (list, numpy 1darray) yang akan ditentukan nilai pdf atau cdf-nya.

loc: bilangan (int, float) yang digunakan sebagai lokasi tengah distribusi normal atau nilai rata-rata data, defaultnya adalah 0.

scale: bilangan (int, float) yang digunakan sebagai sebaran distribusi normal terhadap lokasi tengahnya atau nilai standar deviasi (unbiased) untuk sampel, defaultnya adalah 1.


kita dapat menentukan nilai probabilitas dan probabilitas kumulatif untuk tinggi badan tertentu. Kita akan menentukan nilai ini untuk tinggi badan 165 cm.

```python
from scipy.stats import norm

# Rata-rata dan standar deviasi unbiased tinggi_badan
tb_mean = tinggi_badan.mean()
tb_std = tinggi_badan.std(ddof=1)
# Tentukanlah nilai probabilitas dan probabilitas kumulatif menggunakan 
# distribusi normal untuk tinggi badan, x = 165 cm
x = 165
pdf_x = norm.pdf(x, loc=tb_mean, scale=tb_std)
cdf_x = norm.cdf(x, loc=tb_mean, scale=tb_std)
print("Probabilitas x = %d cm adalah %.4f." % (x, pdf_x))
print("Probabilitas kumulatif x = %d cm adalah %.4f." % (x, cdf_x))
```
